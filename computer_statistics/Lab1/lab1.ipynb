##### List()

```python
import numpy as np
import pandas as pd

# Tạo data frame gồm ba cột time, x, y
time = np.arange(0, 10, 0.1)
x = np.sin(time)
y = np.cos(time)
df = pd.DataFrame({'time': time, 'x': x, 'y': y})
```

```python
print(df) # In ra data frame
```

```python
df.head(3) # 3 dòng đầu
```

```python
df.tail(3) # 3 dòng cuối
```

```python
df['time'] # Lấy cột time
```

```python
df[['time', 'x']] # Lấy cột time và x
```

```python
df[3:5] # Lấy dòng từ 3 đến 5   
```

```python
df.hist() # Vẽ histogram
```
