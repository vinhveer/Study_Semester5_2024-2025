```python
df_Chinstrap = df[df['species'] == 'Chinstrap']
df_Gentoo = df[df['species'] == 'Gentoo']
df_Adelie = df[df['species'] == 'Adelie']

count_chinstrap = df_Chinstrap.count()
count_gentoo = df_Gentoo.count()
count_adelie = df_Adelie.count()


count_chinstrap = count_chinstrap['species']
count_gentoo = count_gentoo['species']
count_adelie = count_adelie['species']

print(count_chinstrap)
print(count_gentoo)
print(count_adelie)
```

```python
# Số lượng chim cánh cụt trên mỗi đảo
count_chinstrap_island = df_Chinstrap['island'].value_counts()
count_adelie_island = df_Adelie['island'].value_counts()
count_gentoo_island = df_Gentoo['island'].value_counts()

# Tạo bảng số lượng chim cánh cụt trên mỗi đảo
island_counts = pd.DataFrame({
    'Chinstrap': count_chinstrap_island,
    'Adelie': count_adelie_island,
    'Gentoo': count_gentoo_island
})

island_counts
```

```python
index = ['mean', 'median', 'std']

df_Cblm = df_Chinstrap['bill_length_mm']
x = [df_Cblm.mean(), df_Cblm.median(), df_Cblm.std()]
df_Cbdm = df_Chinstrap['bill_depth_mm']
x1 = [df_Cbdm.mean(), df_Cbdm.median(), df_Cbdm.std()]
df_Cf = df_Chinstrap['flipper_length_mm']  # Fixed column name
x2 = [df_Cf.mean(), df_Cf.median(), df_Cf.std()]
df_Cbm = df_Chinstrap['body_mass_g']
x3 = [df_Cbm.mean(), df_Cbm.median(), df_Cbm.std()]

df_Chinstrap_result = pd.DataFrame(
    {
        'bill_length_mm': x,
        'bill_depth_mm': x1,
        'flipper_length_mm': x2,
        'body_mass_g': x3
    }, index=index
)

df_Chinstrap_result
```

```python
index = ['mean', 'median', 'std']

df_gentoo_blm = df_Gentoo['bill_length_mm']
y = [df_gentoo_blm.mean(), df_gentoo_blm.median(), df_gentoo_blm.std()]
df_gentoo_bdm = df_Gentoo['bill_depth_mm']
y1 = [df_gentoo_bdm.mean(), df_gentoo_bdm.median(), df_gentoo_bdm.std()]
df_gentoo_f = df_Gentoo['flipper_length_mm']  # Fixed column name
y2 = [df_gentoo_f.mean(), df_gentoo_f.median(), df_gentoo_f.std()]
df_gentoo_bm = df_Gentoo['body_mass_g']
y3 = [df_gentoo_bm.mean(), df_gentoo_bm.median(), df_gentoo_bm.std()]

df_Gentoo_result = pd.DataFrame(
    {
        'bill_length_mm': y,
        'bill_depth_mm': y1,
        'flipper_length_mm': y2,
        'body_mass_g': y3
    }, index=index
)

df_Gentoo_result
```

```python
index = ['mean', 'median', 'std']

df_adelie_blm = df_Adelie['bill_length_mm']
z = [df_adelie_blm.mean(), df_adelie_blm.median(), df_adelie_blm.std()]
df_adelie_bdm = df_Adelie['bill_depth_mm']
z1 = [df_adelie_bdm.mean(), df_adelie_bdm.median(), df_adelie_bdm.std()]
df_adelie_f = df_Adelie['flipper_length_mm']  # Fixed column name
z2 = [df_adelie_f.mean(), df_adelie_f.median(), df_adelie_f.std()]
df_adelie_bm = df_Adelie['body_mass_g']
z3 = [df_adelie_bm.mean(), df_adelie_bm.median(), df_adelie_bm.std()]

df_Adelie_result = pd.DataFrame(
    {
        'bill_length_mm': z,
        'bill_depth_mm': z1,
        'flipper_length_mm': z2,
        'body_mass_g': z3
    }, index=index
)

df_Gentoo_result
```

#### Câu 3. Vẽ đồ thị phù hợp thể hiện phân phối của ba loài chim cánh cụt, qua đó bạn hãy đưa ra bình luận của mình.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Giả sử df là DataFrame chứa dữ liệu của các loài chim cánh cụt
plt.figure(figsize=(15, 10))

# Bill Length
plt.subplot(2, 2, 1)
sns.boxplot(x='species', y='bill_length_mm', data=df)
plt.title('Bill Length Distribution by Species')

# Bill Depth
plt.subplot(2, 2, 2)
sns.boxplot(x='species', y='bill_depth_mm', data=df)
plt.title('Bill Depth Distribution by Species')

# Flipper Length
plt.subplot(2, 2, 3)
sns.boxplot(x='species', y='flipper_length_mm', data=df)
plt.title('Flipper Length Distribution by Species')

# Body Mass
plt.subplot(2, 2, 4)
sns.boxplot(x='species', y='body_mass_g', data=df)
plt.title('Body Mass Distribution by Species')

plt.tight_layout()
plt.show()
```

   - Loài Chinstrap và Gentoo có độ sâu mỏ tương đối tương đương nhau, nhưng vẫn thấp hơn so với loài Adelie.

3. **Flipper Length (Độ dài cánh)**:
   - Loài Gentoo có độ dài cánh lớn nhất, với trung vị cao hơn hẳn so với hai loài còn lại.
   - Loài Chinstrap có độ dài cánh trung bình, nằm giữa loài Gentoo và loài Adelie.

4. **Body Mass (Trọng lượng cơ thể)**:

- Loài Adelie có xu hướng có độ sâu mỏ lớn nhất nhưng lại có các đặc điểm khác nhỏ hơn so với hai loài còn lại.
- Loài Chinstrap có các đặc điểm trung bình, nằm giữa loài Gentoo và loài Adelie.

#### Câu 4. Chọn loại đồ thị phù hợp để khảo sát mối quan hệ giữa độ dài cánh và trọng lượng của loài chim cánh cụt Gentoo

```python
import matplotlib.pyplot as plt

# Giả sử du_lieu là DataFrame chứa dữ liệu của các loài chim cánh cụt
gentoo_data = df[df['species'] == 'Gentoo']

plt.figure(figsize=(10, 6))
plt.scatter(gentoo_data['flipper_length_mm'], gentoo_data['body_mass_g'], alpha=0.5)
plt.title('Biểu đồ phân tán giữa độ dài cánh và trọng lượng của loài Gentoo')
plt.xlabel('Độ dài cánh (mm)')
plt.ylabel('Trọng lượng (g)')
plt.grid(True)
plt.show()
```

- Dựa vào biểu đồ scatter, ta có thể kết luận như sau
    + Xu hướng phân bố dữ liệu: Dữ liệu có xu hướng phân bố theo cùng một hướng, nhưng hơi rời rạc
    + Về tương quan, dữ liệu có xu hướng theo một đường thẳng
- **Kết luận:** Có mội quan hệ giữa các đại lượng này

##### Ngoài độ dài cánh, còn một số đại lượng có thể ảnh hưởng đến trọng lượng của Chim cánh cụt, để kiểm chứng, ta sẽ vẽ biểu đồ scatter để chứng minh

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Giả sử du_lieu là DataFrame chứa dữ liệu của các loài chim cánh cụt
du_lieu_gentoo = df[df['species'] == 'Gentoo']

plt.figure(figsize=(15, 10))

# Độ dài mỏ và trọng lượng
plt.subplot(2, 2, 1)
sns.scatterplot(data=du_lieu_gentoo, x='bill_length_mm', y='body_mass_g')
plt.title('Độ dài mỏ và trọng lượng của loài Gentoo')
plt.xlabel('Độ dài mỏ (mm)')
plt.ylabel('Trọng lượng (g)')

# Độ sâu mỏ và trọng lượng
plt.subplot(2, 2, 2)
sns.scatterplot(data=du_lieu_gentoo, x='bill_depth_mm', y='body_mass_g')
plt.title('Độ sâu mỏ và trọng lượng của loài Gentoo')
plt.xlabel('Độ sâu mỏ (mm)')
plt.ylabel('Trọng lượng (g)')

# Giới tính và trọng lượng
plt.subplot(2, 2, 3)
sns.scatterplot(data=du_lieu_gentoo, x='sex', y='body_mass_g')
plt.title('Giới tính và trọng lượng của loài Gentoo')
plt.xlabel('Giới tính')
plt.ylabel('Trọng lượng (g)')

plt.tight_layout()
plt.show()
```

Dựa vào các biểu đồ, ta thấy xu hướng của các biểu đồ như sau:
- Với độ dài mỏ và trọng lượng của loài Gentoo: Biểu đồ có xu hướng đi theo một đường $\to$ có ảnh hưởng
- Với độ sâu mỏ và trọng lượng của loài Gentoo: Biểu đồ có xu hướng đi theo một đường $\to$ có ảnh hưởng
- Với giới tính và trọng lượng của loài Gentoo: Biểu đồ có xu hướng đi theo hai đường thẳng khác nhau, nhưng female và male có lệch $\to$ có ảnh hưởng
