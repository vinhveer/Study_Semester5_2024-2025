##### Dùng hàm matplotlib.pyplot.hist

```python
import matplotlib.pyplot as plt
import numpy as np

# Tạo dữ liệu ngẫu nhiên
a, b = 0, 30
data = np.random.uniform(0, 30, 100)

n, bins, patches = plt.hist(data, bins='auto', color='#0504aa', alpha=0.9, rwidth=0.8)

plt.grid(axis='y', alpha=0.75)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('My Very Own Histogram')

# Tính trung bình và độ lệch chuẩn cho phân phối đều
mu = (a + b) / 2
sigma = (b - a) / np.sqrt(12)
# Hiển thị μ và σ trên biểu đồ
plt.text(20, 18, r'$\mu=%.2f, \sigma=%.2f$' % (mu, sigma))

maxfreq = n.max()
plt.ylim(ymax=np.ceil(maxfreq/10)*10 if maxfreq % 10 else maxfreq + 10)
```

##### Dataset 15 (Practice)

```python
import pandas as pd
import matplotlib.pyplot as plt

data_import = pd.read_excel('15_OLDFAITH.xls')
df = pd.DataFrame(data_import)

n, bins, patches = plt.hist(df['Duration'], bins='auto', color='#0504aa', alpha=0.9, rwidth=0.8)

plt.grid(axis='y', alpha=0.75)
plt.xlabel('Duration (giây)')
plt.ylabel('Frequency')
plt.title('Histogram of Duration')

# Tính trung bình và độ lệch chuẩn cho phân phối đều
mean_duration = df['Duration'].mean()
std_duration = df['Duration'].std()

# Hiển thị μ và σ trên biểu đồ
plt.text(80, 18, r'$\mu=%.2f, \sigma=%.2f$' % (mean_duration, std_duration))

plt.show()
```

##### Dataset 4: Nhựa/CO trong thuốc lá

```python
import pandas as pd
import matplotlib.pyplot as plt

# Đọc dữ liệu từ tệp Excel
data_frame = pd.read_excel('04_CIGARET.xls')

# Trích xuất các thuộc tính cần thiết từ DataFrame
real_tar_weight = data_frame['KgTar'] 
real_co_weight = data_frame['KgCO']   
mean_tar_weight = data_frame['MnTar'] 
mean_co_weight = data_frame['MnCO']   
low_tar_weight = data_frame['FLTar']  
low_co_weight = data_frame['FLCO']    

# Tạo một figure với 3 subplots
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Biểu đồ 1: KgTar vs KgCO (Thực tế)
axs[0].scatter(real_tar_weight, real_co_weight, color='blue')
axs[0].set_title('KgTar vs KgCO (Real)')
axs[0].set_xlabel('Tar Weight (Kg)')
axs[0].set_ylabel('CO Weight (Kg)')

# Biểu đồ 2: MnTar vs MnCO (Trung bình)
axs[1].scatter(mean_tar_weight, mean_co_weight, color='green')
axs[1].set_title('MnTar vs MnCO (Mean)')
axs[1].set_xlabel('Tar Weight (Kg)')
axs[1].set_ylabel('CO Weight (Kg)')

# Biểu đồ 3: FLTar vs FLCO (Mức thấp)
axs[2].scatter(low_tar_weight, low_co_weight, color='red')
axs[2].set_title('FLTar vs FLCO (Low)')
axs[2].set_xlabel('Tar Weight (Kg)')
axs[2].set_ylabel('CO Weight (Kg)')

# Căn chỉnh khoảng cách giữa các subplots
plt.tight_layout()

# Hiển thị biểu đồ
plt.show()
```

##### **Bài tập 1:** Xây dựng bar char và pie char cho dữ liệu ở bảng sau. So sánh 2 biểu đồ trên, biểu đồ nào là hiệu quả hơn trong việc hiển thị thông tin

```python
import matplotlib.pyplot as plt

# Data from the image
colleges = ['Public 2-Year', 'Public 4-Year', 'Private 2-Year', 'Private 4-Year']
relative_frequencies = [36.8, 40.0, 1.6, 21.9]

# Plotting a bar chart
plt.figure(figsize=(12, 5))

# Bar Chart
plt.subplot(1, 2, 1)
plt.bar(colleges, relative_frequencies, color='skyblue')
plt.title('Bar Chart of College Frequencies')
plt.xlabel('College Type')
plt.ylabel('Relative Frequency (%)')
plt.xticks(rotation=45)

# Pie Chart
plt.subplot(1, 2, 2)
plt.pie(relative_frequencies, labels=colleges, autopct='%1.1f%%', colors=['skyblue', 'orange', 'lightgreen', 'lightcoral'], startangle=140)
plt.title('Pie Chart of College Frequencies')

# Display both charts
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Đọc dữ liệu từ file Excel
df = pd.read_excel('17_COLA.xls')  # Đảm bảo đường dẫn đúng

# Kiểm tra tên cột
print(df.columns)  # Chạy dòng này trước để biết tên chính xác

# Sử dụng tên cột chính xác cho trọng lượng của Diet Coke
# Thay 'CKREGWT' bằng tên cột thực tế chứa trọng lượng của Diet Coke nếu cần
Diet_coke = df['CKREGWT'].dropna()  # Bỏ qua các giá trị null nếu có

# Tạo biểu đồ với kích thước lớn hơn
plt.figure(figsize=(12, 8))  # Điều chỉnh kích thước biểu đồ (width, height)

# Vẽ biểu đồ histogram cho trọng lượng của Diet Coke
n, bins, patches = plt.hist(Diet_coke, bins='auto', color='#0504aa', alpha=0.7, rwidth=0.85)

# Hiển thị các đường lưới
plt.grid(axis='y', alpha=0.75)
plt.xlabel('Trọng lượng (pound)')  # Đặt nhãn trục x cho trọng lượng
plt.ylabel('Tần suất')
plt.title('Histogram trọng lượng của Diet Coke')  # Cập nhật tiêu đề

# Tính trung bình và độ lệch chuẩn
mu = np.mean(Diet_coke)
sigma = np.std(Diet_coke)

# Thêm văn bản cho trung bình và độ lệch chuẩn
plt.text(mu + 1, max(n) * 0.9, r'$\mu=%.2f, \sigma=%.2f$' % (mu, sigma), fontsize=12)

# Giới hạn trục y
plt.ylim(ymax=np.ceil(max(n) / 10) * 10 if max(n) % 10 else max(n) + 10)

# Hiển thị biểu đồ
plt.show()
```
